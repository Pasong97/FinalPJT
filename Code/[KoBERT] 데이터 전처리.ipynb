## (2023.02.07 추가) 데이터 추가

```python
df3 = pd.read_excel('/content/drive/MyDrive/final project/data/KoBERT Data/한국어_연속적_대화_데이터셋.xlsx')
df3
```

```python
df3 = df3.iloc[:,1:]
df3
```

```python
df3.Emotion.unique()
```

```python
## '분노', '혐오' >> '짜증'으로 단일화
Emotion = {'Emotion' : {'분노':'짜증', '행복':'기쁨', '공포':'불안', '놀람':'당황', '슬픔':'슬픔', '중립':'중립', '혐오':'짜증'}}
df3 = df3.replace(Emotion)
df3
```

```python
df3.Emotion.unique()
```

```python
df3.Emotion.value_counts()
```

```python
## 중립 15000개 추출해서 병합할 것임
neutrality = df3[df3['Emotion'] == '중립'][:15000]
neutrality
```

```python
## 당황 4857개 추출해서 병합할 것임
embarrassed = df3[df3['Emotion'] == '당황']
embarrassed
```

```python
## 기쁨 1030개 추출해서 병합할 것임
happiness = df3[df3['Emotion'] == '기쁨']
happiness
```

```python
df = pd.read_csv('/content/drive/MyDrive/final project/data/KoBERT Data/dataset.csv')
df
```

```python
df = df.iloc[:,:2]
df
```

```python
df.Emotion.unique()
```

```python
## '분노', '혐오' >> '짜증'으로 단일화
Emotion = {'Emotion' : {'분노':'짜증', '행복':'기쁨', '공포':'불안', '놀람':'당황', '슬픔':'슬픔', '중립':'중립', '혐오':'짜증'}}
df = df.replace(Emotion)
df
```

```python
df.Emotion.unique()
```

```python
df['Emotion'].value_counts()
```

```python
data = pd.concat([df,neutrality,embarrassed,happiness])
data
```
